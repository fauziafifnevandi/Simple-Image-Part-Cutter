```python
import numpy as np
import cv2
import glob, os

os.chdir("C:/Users/Nyx/skripsi/cnn_fauzi/default_model/slice_bukan_wajah/default/part1")
```

```python
for file in glob.glob("*.jpg"): 
    img = cv2.imread(file)
    #img = img[:,:,0]
    l = img.shape[1]
    t = img.shape[0]
    p = 128
    
    for i in range (0, t//p):
        for j in range (0,l//p):
            nama_output = "output/" + file +"."+ str(i) + "." +str(j)+".jpg"
            cv2.imwrite(nama_output,img[p*i:p*(i+1),p*j:p*(j+1)]) 
            print(nama_output)
    

```
